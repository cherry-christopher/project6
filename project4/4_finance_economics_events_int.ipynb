# Remove Duplicate Rows

````python
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.gdp AS
WITH RankedData AS (
  SELECT *,
         ROW_NUMBER() OVER (
           PARTITION BY
             SAFE_CAST(year AS STRING),
             state,
             category,
             SAFE_CAST(value AS STRING)
           ORDER BY _load_time DESC  -- Keeps the most recent entry
         ) AS row_num
  FROM finance_economics_events_int.gdp
)
SELECT * EXCEPT(row_num)
FROM RankedData
WHERE row_num = 1
ORDER BY year,state;
````

````python
# crime_data
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.crime_data AS
WITH RankedData AS (
  SELECT *,
         ROW_NUMBER() OVER (
           PARTITION BY
             SAFE_CAST(year AS STRING),
             SAFE_CAST(population AS STRING),
             state,
             SAFE_CAST(crime AS STRING),
             SAFE_CAST(crime_against_persons AS STRING),
             SAFE_CAST(crime_against_property AS STRING),
             SAFE_CAST(crime_against_society AS STRING)
           ORDER BY _load_time DESC  -- Keeps the most recent entry
         ) AS row_num
  FROM finance_economics_events_int.crime_data
)
SELECT * EXCEPT(row_num)
FROM RankedData
WHERE row_num = 1
ORDER BY year,state;
````

````python
# mortality_and_natality
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.mortality_and_natality AS
WITH RankedData AS (
  SELECT *,
         ROW_NUMBER() OVER (
           PARTITION BY
             SAFE_CAST(year AS STRING),
             state,
             month,
             period,
             indicator,
             SAFE_CAST(data_value AS STRING)
           ORDER BY _load_time DESC  -- Keeps the most recent entry
         ) AS row_num
  FROM finance_economics_events_int.mortality_and_natality
)
SELECT * EXCEPT(row_num)
FROM RankedData
WHERE row_num = 1
ORDER BY state,year,month DESC;
````

````python
# state_energy_consumption
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.state_energy_consumption AS
WITH RankedData AS (
  SELECT *,
         ROW_NUMBER() OVER (
           PARTITION BY
             state,
             SAFE_CAST(year AS STRING),
             SAFE_CAST(value AS STRING),
             sector
           ORDER BY _load_time DESC  -- Keeps the most recent entry
         ) AS row_num
  FROM finance_economics_events_int.state_energy_consumption
)
SELECT * EXCEPT(row_num)
FROM RankedData
WHERE row_num = 1
order by year, state desc;
````

````python
# us_geocode
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.us_geocode AS
WITH RankedData AS (
  SELECT *,
         ROW_NUMBER() OVER (
           PARTITION BY
             state_full_name
           ORDER BY _load_time DESC  -- Keeps the most recent entry
         ) AS row_num
  FROM finance_economics_events_int.us_geocode
)
SELECT * EXCEPT(row_num)
FROM RankedData
WHERE row_num = 1
order by state_full_name desc;
````

## mortality_and_natality

This table decribes infants and death number by month

This code will aggrate 12 months into 1 year

````python
# mortality_and_natality
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.mortality_and_natality AS
SELECT
  year,
  state,
  indicator,
  SUM(data_value) as data_value,  -- Summing values for all 12 months
  _data_source,
  MAX(_load_time) AS _load_time  -- Keeping the most recent load time
FROM finance_economics_events_int.mortality_and_natality
GROUP BY state, year, indicator, 	_data_source;
````

## state_level_unemployment_data

This table describes state level unemployemnt data by month

This code calculates annual average state levle unemployement number

````python
# state_level_unemployment_data
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.state_level_unemployment_data AS
SELECT
  year,
  state_and_area,
  AVG(unemployment) AS unemployment,  -- Calculate annual average unemployment
  _data_source,
  MAX(_load_time) AS _load_time  -- Keep the most recent load time
FROM finance_economics_events_int.state_level_unemployment_data
GROUP BY year, state_and_area, _data_source
order by year, state_and_area;
````

## gdp


This table has a column "category" which contains 3 values: "personal_income_gdp", "Real gross domestic product", and "Current-dollar gross domestic product"

This table replace "category" with those 3 new columns
Annualize 2023&2024 as we dont have compelete data on these 2 years

````python
%%bigquery
CREATE OR REPLACE TABLE finance_economics_events_int.gdp AS
SELECT
  year,
  state,
  MAX(CASE WHEN category = 'personal_income_gdp' THEN value END) AS personal_income_gdp,
  MAX(CASE WHEN category = 'Real gross domestic product' THEN value END) AS real_gdp,
  MAX(CASE WHEN category = 'Current-dollar gross domestic product' THEN value END) AS current_dollar_gdp,
  _data_source,
  _load_time
FROM finance_economics_events_int.gdp
GROUP BY year, state, _data_source, _load_time;
````

````python
  %%bigquery
  CREATE OR REPLACE TABLE finance_economics_events_int.gdp AS
  SELECT
    CASE
      WHEN year IN (2023.0, 2023.75, 2023.99) THEN 2023.0  -- Replace 2023.5 & 2023.99 with 2023.0
      WHEN year IN (2024.25, 2024.5, 2024.75) THEN 2024.0  -- Replace 2024 quarter entries with 2024.0
      ELSE year  -- Keep other years unchanged
    END AS year,
    state,
    AVG(personal_income_gdp) * 4 AS personal_income_gdp,  -- Compute the annualized average
    AVG(real_gdp) * 4 AS real_gdp,
    AVG(current_dollar_gdp) * 4 AS current_dollar_gdp,
    _data_source,
    MAX(_load_time) AS _load_time
  FROM finance_economics_events_int.gdp
  WHERE year NOT IN (2024.0)  -- Remove any pre-existing 2024.0 to avoid duplicates
  GROUP BY year, state, _data_source
````
